```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
from scipy.io import wavfile
```

Zadatak 1.

```python
#Impulsni odziv idealnog Hilbertovog transformatora
#Kako je impulsni odziv idealnog Hilbertovog transformatora beskonačan i nekauzalan
#Posmatraćemo ga na konačnom intervalu od 0 do 2*N tako da bude konačan
#I zakasnićemo ga u vremenu za N odmjeraka kako bi bio kauzalan
N = 50
tau = 0.0000001 #unosimo i ovo kašnjenje, osim N, kako bismo izbjegli računanje odziva u 0
n = np.arange(0,2*N)
#vrijednost odziva za n različito od 0
h = (2/(np.pi*(n-N - tau)))*((np.sin(0.5*np.pi*(n-N - tau)) ** 2))

w, H = signal.freqz(h, 1)
#H /= np.max(H)

plt.figure()
plt.stem(n, h)
plt.xlabel('n')
plt.ylabel('h(n)')
```

```python
#Crtanje spektra idealnog Hilbertovog transformatora za impulsni odziv h(n)
plt.subplot(3, 1, 1)
plt.title('Spektar impulsnog odziva idealnog Hilbertovog transformatora:')
plt.plot(w/np.pi, np.abs(H))
plt.xlabel(r'$\omega$')
plt.ylabel(r'$\vert H(e^{j \omega}) \vert$')
  
plt.subplot(3, 1, 2)
plt.plot(w/np.pi, 20*np.log10(np.abs(H)))
plt.ylabel('Amplituda [dB]')
plt.xlabel('Frekvencija [rad/odmjerku]')

plt.subplot(3, 1, 3)
angles = np.unwrap(np.angle(H))
plt.plot(w/np.pi, angles)
plt.ylabel('Faza')
plt.xlabel(r'$\omega$/$\pi$')
```

Zadatak 6.

```python
def plot_dtft(x, fs):
    '''Crta amplitudni spektar signala signala x.
       Frekvencijska osa je u Hercima za frekvenciju odmjeravanja fs.
    '''
    
    from scipy.fftpack import rfft, rfftfreq
    
    L = np.int(2**np.ceil(np.log2(len(x))))
    X = rfft(x, L)
    X /= L
    F = rfftfreq(L, 1/fs)
    
    plt.figure()
    plt.plot(F, 20*np.log10(np.abs(X)))
    plt.xlabel('F [Hz]')
    plt.ylabel('|X(F)| [dB]')
    
```

```python
#Generisanje ulaznog signala m(t) =cos(2π800t)+0.5*cos(2π1000t)+2*cos(2π1200t)
Fs = 48e3
Fc = 10e3
n = np.arange(0,Fs)
m = np.cos(2*np.pi*800*n/Fs) + 0.5*np.cos(2*np.pi*1000*n/Fs) + 2*np.cos(2*np.pi*1200*n/Fs)
#Posmatramo najprije gornji bočni opseg(USB) i zbog toga je nosilac dat na sljedeći način:
c = np.exp(2.0j*np.pi*Fc*n/Fs)
```

```python
plt.figure()
plt.stem(m)
plt.xlabel('n')
plt.ylabel('m(n)')
```

Nacrtaćemo ulazni signal na intervalu od 0 do Fs sa pomakom 1000, kako bismo bolje vidjeli generisani signal.

```python
nn = np.arange(0,Fs,1000)
mn = np.cos(2*np.pi*800*nn/Fs) + 0.5*np.cos(2*np.pi*1000*nn/Fs) + 2*np.cos(2*np.pi*1200*nn/Fs)
plt.figure()
plt.stem(mn)
plt.xlabel('n')
plt.ylabel('m(n)')
```

Nacrtajmo spektar ulaznog signala.

```python
plot_dtft(m,Fs)
```

Na amplitudskoj karakteristici ulaznog signala priloženoj iznad uočavamo 3 harmonka na frekvencijama 800, 1000 i 1200 Hz i amplituda 1, 0.5 i 2, respektivno.

Sada ćemo implementirati SSB modulator, ali umjesto gore-realizovanog filtra koristiću metodu signal.hilbert(x) koja vraća analitički signal čiji je realni dio ulazni signal m(n), a imaginarni Hilbertova transformacija ulaznog signala.

Posmatrajmo prvo gornji bočni opseg - USB.

```python
mc=signal.hilbert(m) #Analitički signal, tj. m + j*Hilbert(m)
ssb_modul= mc * c #Modulacija
ssb_modul_real = np.real(ssb_modul) #Izdvajanje realnog dijela signala
#Računanje i crtanje spektra analitičkog signala
w , Mc = signal.freqz(mc, 1)
plt.figure()
plt.plot(Fs*w/np.pi/2, 20*np.log10(np.abs(Mc)))
plt.xlabel('F [Hz]')
plt.ylabel('Amplituda [dB]')
#Crtanje amplitudnog spektra modulisanog signala (realnog dijela signala),tj. gornji bočni opseg(USB)
plot_dtft(ssb_modul_real,Fs)
```

Dio zadatka 11. Implementacija SSB modulatora bez korištenja kompleksne aritmetike - Gornji bočni opseg (USB)

```python
#EFIKASNIJA IMPLEMENTACIJA SSB modulatora - USB: Koristimo sinus i kosinus umjesto množenja signala
#mc(n) kompleksinim signalom c(n)
ssb1 = m * np.cos(2*np.pi*Fc*n/Fs)
ssb2 = np.imag(mc) * np.sin(2*np.pi*Fc*n/Fs) #np.imag(mc) je Hilbertova transformacija ulaznog signala
ssb = ssb1 - ssb2 #ova 2 signala oduzimamo u slučaju USB
```

```python
plot_dtft(ssb,Fs)
```

Vidimo da smo izdvojili gornji bočni opseg (USB), pomjeren desno od frekvencije nosioca (Fc=10kHz).

Posmatrajmo sada donji bočni opseg - LSB.

```python
#Generisanje ulaznog signala m(t) =cos(2π800t)+0.5*cos(2π1000t)+2*cos(2π1200t)
Fs = 48e3
Fc = 10e3
n = np.arange(0,Fs)
m = np.cos(2*np.pi*800*n/Fs) + 0.5*np.cos(2*np.pi*1000*n/Fs) + 2*np.cos(2*np.pi*1200*n/Fs)
#Sada posmatramo donji bočni opseg(LSB) i zbog toga je nosilac dat na sljedeći način:
c = np.exp(-2.0j*np.pi*Fc*n/Fs)
```

```python
mc=signal.hilbert(m) #Analitički signal, tj. m + j*Hilbert(m)
ssb_modul= mc * c #Modulacija
ssb_modul_real = np.real(ssb_modul) #Izdvajanje realnog dijela signala
#Crtanje amplitudnog spektra modulisanog signala (realnog dijela signala),tj. gornji bočni opseg(USB)
plot_dtft(ssb_modul_real,Fs)
```

Vidimo da smo izdvojili donji bočni opseg (LSB), pomjeren lijevo od frekvencije nosioca (Fc=10kHz).

Dio zadatka 11. Implementacija SSB modulatora bez korištenja kompleksne aritmetike - Donji bočni opseg (LSB)

```python
#EFIKASNIJA IMPLEMENTACIJA SSB modulatora - LSB: Koristimo sinus i kosinus umjesto množenja signala
#m(n) kompleksinim signalom c(n)
ssb1 = m * np.cos(2*np.pi*Fc*n/Fs)
ssb2 = np.imag(mc) * np.sin(2*np.pi*Fc*n/Fs) #np.imag(mc) je Hilbertova transformacija ulaznog signala
ssb = ssb1 + ssb2 #ova 2 signala sabiramo u slučaju LSB
```

```python
plot_dtft(ssb,Fs)
```

Zadatak 7.

```python
# Implementacija SSB demodulatora - LSB
m_demodul=signal.hilbert(ssb_modul_real) #analitički signal
ssb_demodul = m_demodul * c
ssb_demodul_real = np.real(ssb_demodul)
#Računanje i crtanje spektra analitičkog signala
w , M_demodul = signal.freqz(m_demodul, 1)
plt.figure()
plt.plot(Fs*w/np.pi/2, 20*np.log10(np.abs(M_demodul)))
plt.xlabel('F [Hz]')
plt.ylabel('Amplituda [dB]')
plot_dtft(ssb_demodul_real,Fs)
```

Na gornjoj amplitudskoj karakteristici nalazi se spektar demodulisanog signala - Donji bočni opseg (LSB).

Dio zadatka 11. Implementacija SSB demodulatora bez korištenja kompleksne aritmetike - Donji bočni opseg (LSB)

```python
#EFIKASNIJA IMPLEMENTACIJA SSB demodulatora - LSB: Koristimo sinus i kosinus umjesto množenja signala
#ssb_demodul_real(n) kompleksinim signalom c(n)
ssb_demodul1 = ssb_modul_real * np.cos(2*np.pi*Fc*n/Fs)
ssb_demodul2 = np.imag(m_demodul) * np.sin(2*np.pi*Fc*n/Fs)
ssb_demoduls = ssb_demodul1 + ssb_demodul2 #ova 2 signala sabiramo u slučaju LSB
```

```python
plot_dtft(ssb_demoduls,Fs)
```

Zadatak 8.

```python
#Analizu i računanje sprovodimo na donjem bočnom opsegu(LSB)
#Generisanje ulaznih signala
#m1(t) = cos(2π800t) + 0.5*cos(2π1000t)+ 2*cos(2π1200t)
#m2(t) = cos(2π1000t) + 0.5*cos(2π1200t) + 2*cos(2π1400t)
Fs = 48e3
Fc1 = 10e3
Fc2 = 20e3
n = np.arange(0,Fs)
m1 = np.cos(2*np.pi*800*n/Fs) + 0.5*np.cos(2*np.pi*1000*n/Fs) + 2*np.cos(2*np.pi*1200*n/Fs)
m2 = np.cos(2*np.pi*1000*n/Fs) + 0.5*np.cos(2*np.pi*1200*n/Fs) + 2*np.cos(2*np.pi*1400*n/Fs)
#Posmatramo donji bočni opseg pa su nosioci dati na sljedeći način:
c1 = np.exp(-2.0j*np.pi*Fc1*n/Fs)
c2= np.exp(-2.0j*np.pi*Fc2*n/Fs)
```

```python
#Računanje ssb modulacije signala m1 i crtanje spektra datog signala
mc1=signal.hilbert(m1) #Analitički signal, tj. m1 + j*Hilbert(m1)
ssb_modul1= mc1 * c1 #Modulacija
ssb_modul_real1 = np.real(ssb_modul1) #Izdvajanje realnog dijela signala
#Crtanje amplitudnog spektra modulisanog signala (realnog dijela signala)
plot_dtft(ssb_modul_real1,Fs)
```

Na amplitudnoj karakteristici iznad nalazi se spektar ssb modulisanog signala m1 nosiocem frekvencije FC1=10kHz.

```python
#Računanje ssb modulacije signala m2 i crtanje spektra datog signala
mc2=signal.hilbert(m2) #Analitički signal, tj. m1 + j*Hilbert(m1)
ssb_modul2= mc2 * c2 #Modulacija
ssb_modul_real2 = np.real(ssb_modul2) #Izdvajanje realnog dijela signala
#Crtanje amplitudnog spektra modulisanog signala (realnog dijela signala)
plot_dtft(ssb_modul_real2,Fs)
```

Na amplitudnoj karakteristici iznad nalazi se spektar ssb modulisanog signala m2 nosiocem frekvencije Fc2=20kHz.

```python
ssb_modul = ssb_modul_real1 + ssb_modul_real2
plot_dtft(ssb_modul,Fs)
```

Na amplitudnoj karakteristici iznad vidimo 2 LSB komponente koje odgovaraju signalima m1 i m2, oko frekvencija nosioca Fc1=10kHz i Fc2=20kHz, respektivno.

Da bismo rekonstruisali pojedine signale prvo iz zbira moramo izdvojiti odgovarajuće kanale, odnosno, opsesge frekvencija. U tu svrhu koristimo filtre propusnike opsega.

```python
plt.figure()
plt.plot(Fs*w/np.pi/2, 20*np.log10(np.abs(H1)))
plt.xlabel(r'F [Hz]')
plt.ylabel(r'$\vert H(F) \vert$')
```

```python
ssb_modul_hat1 = signal.lfilter(h1, 1, ssb_modul)

plot_dtft(ssb_modul_hat1, Fs)
```

Na osnovu spektra filtriranog signala jasno je da smo izdvojili komponentu spektra oko frekvencije prvog nosioca, Fc1=10kHz.

Izvršimo sada SSB demodulaciju na osnovu našeg SSB demodulatora.

```python
m_demodul_hat1=signal.hilbert(ssb_modul_hat1)
ssb_demodul_hat1 = m_demodul_hat1 * c1
ssb_demodul_real_hat1 = np.real(ssb_demodul_hat1)
plot_dtft(ssb_demodul_real_hat1,Fs)
```

Na osnovu reprodukovanog signala možemo zaključiti da smo uspješno rekonstruisali ulazni govorni signal iz fajla "tri.wav".

```python
def delay(signal,N):
    d=np.zeros((1,N+1))
    signal=np.append(d,signal)
    return signal
```

```python
#Generisanje ulaznog signala m(t) =cos(2π800t)+0.5*cos(2π1000t)+2*cos(2π1200t)
Fs = 48e3
Fc = 10e3
n = np.arange(0,Fs)
m = np.cos(2*np.pi*800*n/Fs) + 0.5*np.cos(2*np.pi*1000*n/Fs) + 2*np.cos(2*np.pi*1200*n/Fs)
#Posmatramo najprije gornji bočni opseg(USB) i zbog toga je nosilac dat na sljedeći način:
c = np.exp(2.0j*np.pi*Fc*n/Fs)
#IMPLEMENTACIJA SSB MODULATORA
Delay = 75 #potrebno kašnjenje ulaznog realnog signala
#Delay = N/2, gdje je N = 151 - red projektovanog filtra u zadatku 2
m_delayed= delay(m, Delay)
m_delayedd = np.resize(m_delayed, m_delayed.size -76) #"skraćivanje" zakašnjenog signala za 76 odmjeraka
m_tilde = signal.lfilter(hr, 1, m) #vrjednost ulaznog signala propuštenog kroz filtar
mc = ( m_delayedd + 1j*m_tilde)
ssb_mod = mc * c
ssb_mod_real = np.real(ssb_mod)
plot_dtft(ssb_mod_real, Fs)
```
